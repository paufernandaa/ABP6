## Problema de investigación
El presente estudio busca analizar si existe una relación entre el consumo de cafeína y el nivel de energía diaria en estudiantes.

## Hipótesis
- Hipótesis nula (H0): No existe relación entre el consumo de cafeína y el nivel de energía.
- Hipótesis alternativa (H1): Existe una relación entre el consumo de cafeína y el nivel de energía.

## Variables

- Nivel de energía diaria: variable cuantitativa.
- Género: variable cualitativa.

## Diseño del estudio
Se plantea un estudio observacional con datos simulados de 100 estudiantes, donde se analizan distintas variables relacionadas con hábitos y estado físico.

# Lección 2: Probabilidad y estadística

## Eventos aleatorios
Se definen los siguientes eventos:
- Evento A: El estudiante consume cafeína.
- Evento B: El estudiante presenta un nivel de energía alto.

## Árbol de probabilidad
Se representa de forma conceptual la relación entre los eventos:

- P(A) = 0.6 (probabilidad de consumir cafeína)
- P(B | A) = 0.7 (probabilidad de tener alta energía dado que consume cafeína)
- P(B | A') = 0.4 (probabilidad de tener alta energía sin consumir cafeína)

Este modelo permite visualizar cómo cambia la probabilidad de energía según el consumo de cafeína.

## Tipo de muestreo
Se utiliza un muestreo aleatorio simple, ya que todos los individuos tienen la misma probabilidad de ser seleccionados.

## Diseño muestral
Se trabaja con un dataset simulado de 100 estudiantes, considerando variables como consumo de cafeína, nivel de energía, horas de sueño, nivel de estrés, entre otras.

## Probabilidades básicas

- Intersección:
P(A ∩ B) = P(A) × P(B | A) = 0.6 × 0.7 = 0.42

- Unión:
P(A ∪ B) = P(A) + P(B) - P(A ∩ B)

- Complementario:
P(A') = 1 - P(A) = 0.4

Estos cálculos permiten entender la relación entre eventos y su ocurrencia conjunta.

# Lección 4: Inferencia estadística

Se realiza un test de hipótesis para evaluar si el nivel de energía promedio de los estudiantes es diferente de 5.

```python
from scipy import stats

# promedio de la muestra
media = df['energia_dia'].mean()

# test t de una muestra (comparando con 5)
t_stat, p_valor = stats.ttest_1samp(df['energia_dia'], 5)

print("Media:", media)
print("t-stat:", t_stat)
print("p-valor:", p_valor)
```

Se realizó un test de hipótesis para evaluar si el nivel de energía promedio es diferente de 5.

El valor p obtenido (0.919) es mayor a 0.05, por lo que no se rechaza la hipótesis nula.

Esto indica que no hay evidencia suficiente para afirmar que el nivel de energía promedio difiere significativamente de 5 en la muestra analizada.

```python
# intervalo de confianza 95%
conf_int = stats.t.interval(
    0.95,
    len(df['energia_dia'])-1,
    loc=media,
    scale=stats.sem(df['energia_dia'])
)

print("Intervalo de confianza:", conf_int)
```

El intervalo de confianza al 95% para el nivel de energía se encuentra entre 4.44 y 5.62.

Esto indica que el promedio poblacional probablemente se encuentra dentro de este rango.

Dado que el valor 5 está dentro del intervalo, se refuerza la conclusión de que no existen diferencias significativas en el nivel de energía promedio.

En base al análisis realizado se puede contemplar que el nivel de energía de los estudiantes tiene una distribución más o menos homogénea. No se encontró una asociación intensa entre el consumo de café y el nivel de energía. Y el test de hipótesis así como el intervalo de confianza apuntan a una media del nivel de energía no significativamente diferente a 5. En suma, todos estos resultados evidencian que en esta muestra la cafeína no tiene una fuerte influencia sobre el nivel de energía.

# Lección 5: Intervalos de confianza

Se calculan intervalos de confianza para estimar el valor promedio de las variables del dataset con distintos niveles de confianza.

```python
# Intervalos de confianza para energia_dia

import numpy as np
from scipy import stats
```

```python
data = df['energia_dia']
n = len(data)
mean = np.mean(data)
sem = stats.sem(data)
```

```python
# 90%
ic_90 = stats.t.interval(0.90, n-1, loc=mean, scale=sem)

# 95%
ic_95 = stats.t.interval(0.95, n-1, loc=mean, scale=sem)

# 99%
ic_99 = stats.t.interval(0.99, n-1, loc=mean, scale=sem)

print("IC 90%:", ic_90)
print("IC 95%:", ic_95)
print("IC 99%:", ic_99)
```

### Interpretación de intervalos de confianza

Al comparar los intervalos de confianza, se observa que a mayor nivel de confianza, más amplio es el intervalo.

El intervalo al 90% es más estrecho, mientras que el de 99% es más amplio, lo que refleja mayor incertidumbre pero mayor seguridad de contener el valor real.

Esto demuestra que el tamaño del intervalo depende del nivel de confianza elegido.

## Lección 6: Test de significancia

```python
from scipy import stats
```

```python
data = df['energia_dia']

# Hipótesis:
# H0: la media = 5
# H1: la media ≠ 5

t_stat, p_value = stats.ttest_1samp(data, 5)

print("Estadístico t:", t_stat)
print("Valor p:", p_value)
```

### Interpretación del test

Se realiza un test de hipótesis para evaluar si el promedio del nivel de energía es diferente de 5.

El valor p obtenido se compara con un nivel de significancia α = 0.05.

Si el valor p es mayor que 0.05, no se rechaza la hipótesis nula, lo que indica que no hay evidencia suficiente para afirmar que el promedio es diferente de 5.

En este caso, el resultado sugiere que el nivel de energía promedio no es significativamente distinto de 5.

### Errores tipo I y II

Error tipo I: Rechazar la hipótesis nula cuando en realidad es verdadera.

Error tipo II: No rechazar la hipótesis nula cuando en realidad es falsa.

### Conclusión

En base al test de hipótesis realizado, no se encontró evidencia estadísticamente significativa para afirmar que el nivel de energía promedio difiere de 5.

Esto refuerza los resultados obtenidos previamente con los intervalos de confianza.
